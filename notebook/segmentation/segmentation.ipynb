import package

```python
from si4ul import segmentation_si
```

Local white segmentation algorithm. If bias is bigger than 1, algorithm can detect black object. If bias is smaller than 1, algorithm can detect white object.
Please prepare your image and input to `segmentation_si.thresholding`.

```python
seg = segmentation_si.thresholding('./image/*.jpg', './image/result')
seg
```

post segmentation inference for local white segmentation

```python
segmentation_si.psegi_thresholding(seg)
```

plot histogram of pixel values in input image

```python
segmentation_si.plot_histogram(seg)
```

plot histogram of pixel values per region

```python
segmentation_si.plot_histogram_region(seg, display_statistics=True)
```
